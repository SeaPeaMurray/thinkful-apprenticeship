# Analytics and experimentation
### Investigating kickstarter data

## Are there superficial characteristics about kickstarters that improve success?
This project is going to explore a kickstarter dataset in primarily two ways:

##### Does name length correlate with kickstarter success?

##### Are some categories of kickstarters more successful than others?


With some hints about these questions we can design an experiment that will verify if some categories statistically perform better than others or if some name lengths are statistically better for starting the kickstarter.

#### Setup
We will be using the standard pandas, matplotlib, seaborn, and other statistical tools to take a look at our data first. We drop 'ID' and other unnecessary columns from our dataframe.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
# sns.set_style('dark')
plt.style.use('ggplot')
%matplotlib inline

kick1 = pd.read_csv('../../../kickstarter_set/ks-projects-201801.csv')
kick1.drop(columns=['ID', 'goal', 'currency', 'pledged'], inplace=True)
kick1.loc[:,['deadline', 'launched']] = kick1[['deadline', 'launched']].apply(pd.to_datetime)
kick1.head(3)
```

```python
kick1.info(), kick1.describe()
```

### Pairplot
The pairplot gives us a high-level view of the data. We can look for odd trends or other things that might stand out.

The plot shows an interesting cling to the axis across some values - I haven't exactly figured that out yet.

```python
pp = sns.pairplot(kick1)
pp.fig.suptitle('Pairplot of kickstarter backers and USD')
```

### Jointplot
This jointplot will give us an idea of all of the points with name length versus actual pledged value relationships. It does not appear that, other than outliers, there is much of a relationship here.

For the y axis, we are using actual pledged versus goal to indicate whether or not the kickstarter was successful and by what degree.

```python
joint = sns.jointplot(x=kick1.name.apply(lambda x: len(str(x))), 
              y=(kick1['usd_pledged_real'] / kick1['usd_goal_real']), alpha=0.35)
joint.fig.set_figwidth(16)
joint.fig.set_figheight(12)
joint.fig.suptitle('Kickstarter name length versus actual pledged USD value', size=20)
```

### Verification
To statistically evalute the claim of little significance of name length, let's run an arbitrary t-test against names with 40 or fewer characters - this subset take sup more than 15 thousand entries, so it is significant enough.

It turns out the pvalue is quite high, indicating any difference is due to random chance.

#### (Side note)
It is very interesting how dips in name lengths exist around multiples of 20. There could exist a price-point difference at these lengths.

```python
kick1['name_length'] = kick1.name.apply(lambda x: len(str(x)))
print((kick1['usd_pledged_real'] / kick1['usd_goal_real']).describe())
print('\nThe number of kickstarter names with less than or equal to 40 characters: {}'
      .format(len(kick1[kick1.name_length >= 40])))
stats.ttest_ind(kick1[kick1.name_length >= 40]['usd_pledged_real'], 
                kick1[kick1.name_length >40]['usd_pledged_real'])
```

### Boxplots across categories
Below we have boxplots of the ratios of actual pledged USD against USD goals by main category.

Once again, other than the outliers, it does not appear that some categories are significantly more successful than others. We can run another test to check the numbers.

```python
fig, ax1 = plt.subplots(1, 1, figsize=(16,12))
sns.boxplot(data=kick1, x='main_category', 
            y=kick1['usd_pledged_real'] / kick1['usd_goal_real'], ax=ax1)
ax1.set_title('Ratio of actual pledged versus goal per main category', size=20)
ax1.set_xlabel('Main category', size=14)
ax1.set_ylabel('Ratio of actual pledged versus goal', size=14)
```

```python
music1 = kick1[kick1.main_category == 'Music']
games1 = kick1[kick1.main_category == 'Games']
tech1 = kick1[kick1.main_category == 'Technology']

stats.f_oneway(music1['usd_pledged_real'] / music1['usd_goal_real'], 
               games1['usd_pledged_real'] / games1['usd_goal_real'], 
               tech1['usd_pledged_real'] / tech1['usd_goal_real'])
```

To check three together that visually appear different, we will use an F-one way test - this can be interpreted very similarly to a t-test. We select 'Music', 'Games', and 'Technology' to verify. This also returns a p-value that indicates the differences are due to random chance.

#### Experimental design

I believe any experiment to further verify these results is tricky, as it requires an investment of long periods of time and effort, but let's tackle one of them - the performances of different categories.

#### Categorical performance test
We can assign five teams of our kickstarter employees two different categories of kickstarts each. The two will be quite unrelated as not to create overlap and association bias; for example, two separate kickstarts for one team in the games and food categories. This will help separate any variables associated with the successes, abilities, and networks of our employees and teams. Their game and food kickstarts will now have the same efforts, networks, and other skills shared. 

The projects will also have the same USD goals, in the same city, and be advertised on the same platforms with the same funding for the same length of time, maybe about one month. If we can create nearly all things equal, we can determine how successful some categories are - *all things equal*.

*If successful,* statistical results will yield significant p-values between different segments below the threshold of 0.05. It is a small sample for a test, but when all things are equal across the kickstarters, we can be confident in our results. To improve this experiment, I would certainly recommend many more groups running the pairs of kickstarters, perhaps closer to 50 teams - but that might be overly burdensome.
