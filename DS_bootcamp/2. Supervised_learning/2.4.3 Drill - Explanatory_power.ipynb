## Drill: Explanatory power - assumptions of linear regression

```python
import math
import warnings

from IPython.display import display
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import linear_model
import statsmodels.formula.api as smf

# Display preferences.
%matplotlib inline
pd.options.display.float_format = '{:.4f}'.format
plt.style.use('ggplot')
```

```python
# Acquire, load, and preview the data.
data = pd.read_csv('https://tf-curricula-prod.s3.amazonaws.com/data-science/Advertising.csv')
data.drop(columns='Unnamed: 0', inplace=True)
display(data.head())
data2 = data.copy()

# Instantiate and fit our model.
regr = linear_model.LinearRegression()
Y = data['Sales'].values.reshape(-1, 1)
X = data[['TV','Radio','Newspaper']]
regr.fit(X, Y)

# Inspect the results.
print('\nCoefficients: \n', regr.coef_)
print('\nIntercept: \n', regr.intercept_)
print('\nR-squared:')
print(regr.score(X, Y))
```

### Predictions

Let's run some predictions and observe the distribution of the residuals.

```python
actual = data['Sales']
predicted = regr.predict(X).ravel()

sns.distplot(actual - predicted)
```

### Linear relationships

We want to know if the features correlate to the outcomes in a linear fashion. The pairplot helps us do this quickly.

```python
sns.pairplot(data)
```

```python
# Newspaper against sales does not seem to behave in a linear fashion
sns.scatterplot(data.Sales, np.log(data['Newspaper']) * 2)
```

```python
# I feel this transformation gives us a more linear relationship
sns.scatterplot(data.Sales, np.sqrt(data.Radio))
```

### Homoscedasticity

What does the scatter plot of predictions versus residuals tell us about the model?

```python
actual = data['Sales']
predicted = regr.predict(X).ravel()

fig, ax1 = plt.subplots(1, 1, figsize=(16,12))
scat = sns.scatterplot(predicted, actual - predicted, ax=ax1)
scat.axhline(y=0)
plt.show()
```

### Imbalance

We are not seeing an even distribution of predictions against residuals - the middle and the outsides of the plot are not balanced.

### Transformations

Let's throw a couple of transformations into the model - converting Newspaper and Radio into something more linear.

```python
data['Newspaper_tran'] = np.log(data.Newspaper)
data['Radio_tran'] = np.sqrt(data.Radio)
# data['TV_tran'] = np.sqrt(data.TV)

# Instantiate and fit our new model.
regr2 = linear_model.LinearRegression()
Y2 = data['Sales'].values.reshape(-1, 1)
X2 = data[['TV', 'Radio_tran', 'Newspaper_tran']]
regr2.fit(X2, Y2)
```

```python
sns.distplot(data.Sales - regr2.predict(X2).ravel())
```

#### Notes: Transformations

Performing some feature transformation on the Newspaper and Radio features was able to provide us with a much more normal distribution of residuals.

```python
fig, ax1 = plt.subplots(1, 1, figsize=(16,12))
scat = sns.scatterplot(regr2.predict(X2).ravel(), data.Sales - regr2.predict(X2).ravel(), ax=ax1)
scat.axhline(y=0)
```

#### Notes: correlation scatterplot

These changes don't seem to yield significant improvements, although it appears to be closer to the mark.

```python
print('\nCoefficients: \n', regr.coef_)
print('\nIntercept: \n', regr.intercept_)
print('\nR-squared:')
print(regr.score(X, Y))

print('\nCoefficients: \n', regr2.coef_)
print('\nIntercept: \n', regr2.intercept_)
print('\nR-squared:')
print(regr2.score(X2, Y2))
```

```python
display(X.corr())
display(X2.corr())
```

## Summary

We have been able to produce a better distribution of the residuals through some minor feature transformations and might have a better error distribution. The correlation matrix indicates some stronger feature independence has been introduced. These changes did not, however yield a better R squared value.
