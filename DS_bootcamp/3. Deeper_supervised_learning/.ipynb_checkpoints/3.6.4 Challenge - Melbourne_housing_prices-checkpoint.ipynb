Using this Kaggle data create a model to predict a house's value. We want to be able to understand what creates value in a house, as though we were a real estate developer.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, Normalizer, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, SGDRegressor, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
sns.set_style('darkgrid')
%matplotlib inline
```

```python
mel_less = pd.read_csv('../../../Supervised_lesson6/Melbourne_housing/MELBOURNE_HOUSE_PRICES_LESS.csv')
mel_full = pd.read_csv('../../../Supervised_lesson6/Melbourne_housing/Melbourne_housing_FULL.csv')
```

```python
mel_less.drop_duplicates(inplace=True)
```

```python
plt.subplots(figsize=(16,9))
sns.distplot(mel_less.Price.dropna(), kde=True)
```

```python
mel_less = mel_less.dropna(subset=['Price'])
```

```python
sns.pairplot(mel_less.drop('Postcode', axis=1))
```

```python
for col in mel_less.columns:
    if col != 'Price':
        print(col, ":", len(mel_less[col].unique()))
```

```python
mel_less.Address.str.split().str[0].str.contains(r'[/]').value_counts()
```

```python
mel_less.Address.str.split().str[0].str.contains(r'[A-Za-z]').value_counts()
```

```python
mel_less['Address_slash'] = mel_less.Address.str.split().str[0].str.contains(r'[/]')
mel_less['Address_unit'] = mel_less.Address.str.split().str[0].str.contains(r'[A-Za-z]')

slash = mel_less[mel_less['Address_slash'] == True].index
unit = mel_less[mel_less['Address_unit'] == True].index
reg = mel_less[(mel_less['Address_slash'] != True) & (mel_less['Address_unit'] != True)].index

new = pd.DataFrame(columns=['Price', 'Addy'])
new.Price = mel_less.Price
new.loc[slash,'Addy'] = 'slash'
new.loc[unit,'Addy'] = 'unit'
new.loc[reg,'Addy'] = 'reg'

fig, ax = plt.subplots(figsize=(16,9))
sns.boxplot(x='Addy', y='Price', data=new)
```

```python
sns.distplot(mel_less.Postcode.value_counts(), kde=False)
```

```python
# mel_less.Postcode.value_counts()[:50]
```

```python
regions = pd.get_dummies(mel_less.Regionname)
methods = pd.get_dummies(mel_less.Method)
types = pd.get_dummies(mel_less.Type)
councils = pd.get_dummies(mel_less.CouncilArea)
df = pd.concat([mel_less, regions, methods, types, councils], axis=1)
df.drop(['Type', 'Regionname', 'Address', 'Method', 'Suburb', 'CouncilArea', 'SellerG', 'Postcode', 'Date'], axis=1, inplace=True)
```

```python
df.head()
```

```python
X = df.drop('Price', axis=1)
y = df['Price']
scaler = StandardScaler(with_mean=False)
norm = Normalizer()
X_s = scaler.fit_transform(X)
X_n = norm.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_n, y)
```

```python
clf_lin = LinearRegression(normalize=False)
clf_sgd = SGDRegressor()
clf_las = Lasso()
clf_rf = RandomForestRegressor()

clf_lin.fit(X_train, y_train)
clf_sgd.fit(X_train, y_train)
clf_las.fit(X_train, y_train)
clf_rf.fit(X_train, y_train)
```

```python
for clf in [clf_lin, clf_sgd, clf_las, clf_rf]:
    print(clf.__class__, clf.score(X_test, y_test))
```

```python
p_f = PolynomialFeatures(degree=2, interaction_only=True)
pipe = Pipeline([('polynomial_features', p_f), ('linear_regression', clf_lin)])
pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```

```python
clf_rf2 = RandomForestRegressor(n_estimators=200, max_features='log2')
clf_rf2.fit(X_train, y_train)
clf_rf2.score(X_test, y_test)
```

```python
y_pred = clf_rf2.predict(X_test)
plt.subplots(figsize=(16,9))
sns.scatterplot(y_test, y_pred, alpha=0.5)
plt.suptitle('Actuals vs. Predictions: Prices', size=18)
```

```python
plt.subplots(figsize=(16,9))
sns.scatterplot(y_test, y_pred-y_test, alpha=0.5)
plt.suptitle('Initial residuals', size=18)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(16,9))
sns.distplot(np.sqrt(df.Distance + 1), ax=ax1)
sns.distplot(np.sqrt(df.Propertycount + 1), ax=ax2)
sns.distplot(np.sqrt(df.Rooms), kde=False, ax=ax3)
```

```python
df.Propertycount = np.sqrt(df.Propertycount)
df.Distance = np.sqrt(df.Distance)
df.Rooms = np.log(df.Rooms)
X = norm.fit_transform(df.drop('Price', axis=1))
y = df.Price
X_train, X_test, y_train, y_test = train_test_split(X, y)
clf_rf2.fit(X_train, y_train)
clf_rf2.score(X_test, y_test)
```
