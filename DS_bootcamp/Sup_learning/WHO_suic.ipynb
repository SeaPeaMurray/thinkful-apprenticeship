```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
filename = 'https://raw.githubusercontent.com/SeaPeaMurray/thinkful/master/_Datasets/who_suicide_statistics.csv'
plt.style.use('ggplot')

# for col in suic.keys():
#   print(suic[col].value_counts())
```

```python
suic = pd.read_csv(filename)
suic = suic.dropna()
suic.head()
```

```python
suic.year.value_counts()
```

```python
pd.get_dummies(suic.age)
```

```python
suic.info()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16,12))
sns.distplot(((suic.suicides_no / suic.population).dropna() + 1).apply(np.log), ax=ax)
```

```python
suic.groupby('year')['suicides_no'].sum().plot(figsize=(16,12))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16,12))
sns.lineplot(data=suic.groupby('year')['suicides_no'].sum(), ax=ax)
```
