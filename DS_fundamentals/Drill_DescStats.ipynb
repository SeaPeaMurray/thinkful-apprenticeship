```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# distvar = np.random.distname()
def plot_dist(dist):
    plt.hist(dist)
    plt.axvline(np.mean(dist), linestyle='solid', color='black')
    plt.axvline(np.mean(dist) - np.std(dist), linestyle='dashed', color='black')
    plt.axvline(np.mean(dist) + np.std(dist), linestyle='dashed', color='black')
    plt.show()
```

```python
normvar = np.random.normal(5, 1.0, 100)
plot_dist(normvar)
```

## Notes

The normal distribution indeed has useful information provided by the mean and standard deviation. It describes the central tendency and the confidence interval of 68 percent reliably and mostly symmetrically. 

```python
betavar = np.random.beta(1, 4, 100)
plot_dist(betavar)
```

## Notes

The central tendency is telling in this distribution well and probably more valuable than the standard deviation. The band is not centered here and as it has an extreme right skew, the assumptions of standard deviations and normal distributions do not apply.

```python
multivar = np.random.multinomial(100, [1/6]*6, size=2)
plot_dist(multivar)
```

## Notes

The distribution here was at first confusing, but upon printing out the dataset, it became more clear what happened. The arrays are already distributions - specifically they are the count values of events. A correction to the code is below.

```python
multivar2 = np.random.multinomial(100, [1/6.]*6, size=100)
plot_dist(multivar2)
```

## Notes

This version of the multinomial tells us much more. It more closely aligns with properties of normal distributions, meaning we can deduce very similar things when applying central tendency and standard deviation techniques.

```python
gumvar = np.random.gumbel(5, 1.0, 100)
plot_dist(gumvar)
```

## Notes 

The gumbel distribution is also not normal, so the standard deviation won't tell us too much. However, we can see what the mean of the distribution is.

```python
chivar = np.random.chisquare(2, 100)
plot_dist(chivar)
```

## Notes

For the chisquare distribution, the mean is still fairly valuable - it tells us central tendency and appears not to be affected by extreme outliers. Importantly though, we can see that the standard distribution gives us very little information, as it is too different from Gaussian.

```python
poissvar = np.random.poisson(5, 100)
plot_dist(poissvar)
```

## Notes

Poisson distributions are not normal, so the intuition of standard deviation does not quite apply - although it doesn't appear to be so different that it offers *zero* information about spread. The mean still seems to be valuable, as extreme outliers are absent.
